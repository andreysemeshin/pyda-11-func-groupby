```python
# Задание 1. Классификация фильмов по рейтингу

import pandas as pd

def rate(row):
    if row['rating'] <= 2:
        return 'Низкий рейтинг'
    elif row['rating'] <= 4:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
    

ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

ratings['class'] = ratings.apply(rate, axis=1)

db_new = ratings.merge(movies, on='movieId', how='left')

db_new[['movieId', 'title', 'rating', 'class']].head()


```

```python
# Задание 2 

geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def geo_class(row):
    for region, cities in geo_data.items():
        if row['keyword'] in cities:
            return region
    return 'undefined'    


search = pd.read_csv('keywords.csv')


search['region'] = search.apply(geo_class, axis=1)

search.sort_values(by='region', ascending=False)  # проверка



```

```python
# Задание 3 (бонус)  # регулярные выражения
import re
import pandas as pd



ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')

years = [i for i in range(1950, 2020)]

def production_year(row):
    pattern = '[\d+]{4}'
    try:
        numb = int(re.findall(pattern, row['title'])[0])
        if numb in years:
            return numb
        else:
            return 1900
    except:
        return 1900
movies['year'] = movies.apply(production_year, axis=1)


db_new = ratings.merge(movies, on='movieId', how='left')[['movieId', 'title', 'rating', 'year']]

db_new[['year', 'rating']].groupby('year').mean().sort_values(by='rating', ascending=False)


```
